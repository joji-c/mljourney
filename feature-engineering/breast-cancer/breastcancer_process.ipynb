```python
X=df.iloc[:,1:-1]
Y=df.iloc[:,-1]
```

```python
df.isnull().sum()
#if there was any null value use simpleimputer to fix them
from sklearn.impute import SimpleImputer
impurter=SimpleImputer(missing_values=np.nan,strategy="mean")
X.iloc[:,1:31]=impurter.fit_transform(X.iloc[:,1:31])
X.head()
```

```python
#no category in x if there is use onehotencoder
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
Y=le.fit_transform(Y)
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=1)
X_train.head()
```
