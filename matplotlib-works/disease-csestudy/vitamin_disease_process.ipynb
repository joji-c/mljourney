```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("C:\\Users\\jojic\\OneDrive\\Desktop\\mljourney\\matplotlib-works\\disease-csestudy\\vitamin_deficiency_disease_dataset.csv")
print(df.head())
print(df.columns)
df.isnull().sum()
```

```python
df["alcohol_consumption"].fillna(df["alcohol_consumption"].mode()[0],inplace=True)
df["symptoms_list"].fillna(df["symptoms_list"].mode()[0],inplace=True)
df.isnull().sum()
```

```python
age_gen_avg=df.groupby("gender")["age"].mean()
plt.bar(age_gen_avg.index,age_gen_avg.values)
```

```python
exer_bmi=df.groupby("exercise_level")["bmi"].mean()
plt.plot(exer_bmi)
```

```python
alcho=df.groupby("alcohol_consumption")["age"].mean()
plt.hist(alcho)
```

```python
smo=df.groupby("smoking_status")["age"].mean()
plt.pie(smo,labels=smo.index,autopct="%1.1f%%")
```
