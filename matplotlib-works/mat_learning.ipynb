```python
courses=["django","datasciense","testing","mearn"]
jan_head_count=[10,19,10,20]
feb_head_count=[11,13,15,22]
plt.title("monthly join count",color="red",size=44,weight="bold")
plt.xlabel("courses",color="blue",size=11)
plt.ylabel("head_count",color="brown",size=22,weight="bold")
plt.plot(courses,jan_head_count,marker="o",linestyle="dashdot",color="yellow",linewidth=5)
plt.plot(courses,feb_head_count,marker="*",linestyle="dashed",color="green",linewidth=1)
plt.show()
```
