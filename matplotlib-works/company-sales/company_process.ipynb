```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("C:\\Users\\jojic\\OneDrive\\Desktop\\mljourney\\matplotlib-works\\company-sales\\company_sales_data.csv")
print(df.head())
print(df.columns)
df.isnull().sum()
```

```python
monthly_profit=df.groupby("month_number")["total_profit"].sum()
plt.plot(monthly_profit,marker="o",linestyle="dotted",color="red",linewidth=3,label="monthly report")
plt.xlabel("Month Number")
plt.ylabel("Sold units number")
plt.legend()
monthly_profit 
```

```python
plt.plot(df["month_number"],df["facecream"],marker="s",markerfacecolor="green",label="facecream")
plt.plot(df["month_number"],df["facewash"],marker="s",markerfacecolor="green",label="facewash")
plt.plot(df["month_number"],df["toothpaste"],marker="s",markerfacecolor="green",label="toothpaste")
plt.plot(df["month_number"],df["bathingsoap"],marker="s",markerfacecolor="green",label="bathingsoap")
plt.plot(df["month_number"],df["shampoo"],color="green",marker="s",markerfacecolor="green",label="shampoo")
plt.plot(df["month_number"],df["moisturizer"],marker="s",markerfacecolor="green",label="moisturizer")
plt.legend()
```

```python
plt.scatter(df["month_number"],df["toothpaste"])
```

```python
cream=df.groupby("month_number")["facecream"].mean()
wash=df.groupby("month_number")["facewash"].mean()
plt.bar(cream.index,cream.values,label="facecream")
plt.bar(wash.index,wash.values,label="facewash")
plt.legend()
plt.grid()
```

```python
plt.hist(df["total_profit"],edgecolor="white")
```

```python
products=["facecream","facewash","toothpaste","bathingsoap","shampoo","moisturizer"]
summary_pie={p:df[p].sum() for p in products}
plt.pie(summary_pie.values(),labels=summary_pie,autopct="%1.1f%%")
```
