```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("C:\\Users\\jojic\\OneDrive\\Desktop\\mljourney\\matplotlib-works\\earthquake-casestudy\\earthquake.csv")
print(df.head())
print(df.columns)
df.isnull().sum()
```

```python
df.fillna({"depth":df["depth"].mean()},inplace=True)
df.fillna({"nst":df["nst"].mean()},inplace=True)
df.fillna({"gap":df["gap"].mean()},inplace=True)
df.fillna({"dmin":df["dmin"].mean()},inplace=True)
df.fillna({"rms":df["rms"].mean()},inplace=True)
df.fillna({"place":"unknown"},inplace=True)
df.fillna({"horizontalError":df["horizontalError"].mean()},inplace=True)
df.fillna({"depthError":df["depthError"].mean()},inplace=True)
df.fillna({"magError":df["magError"].mean()},inplace=True)
df.fillna({"magNst":df["magNst"].mean()},inplace=True)

df.isnull().sum()
```

```python
type_status=df["status"].value_counts()
print(type_status)
plt.plot(type_status)
```

```python
mag_type=df.groupby("type")["mag"].sum().head()
plt.plot(mag_type)
```

```python
lat_avg=df.groupby("type")["latitude"].mean()
plt.plot(lat_avg)
```

```python
lon_avg=df.groupby("type")["longitude"].mean()
plt.plot(lon_avg)
```

```python
dep_avg=df.groupby("type")["depth"].mean()
plt.plot(dep_avg)
```
