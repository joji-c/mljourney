```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("C:\\Users\\jojic\\OneDrive\\Desktop\\mljourney\\matplotlib-works\\fraud-case\\Fraud Detection Transactions Dataset.csv") 
print(df.head())
print(df.columns)
df.isnull().sum()
```

```python
tr_fr=df.groupby("Transaction_Type")["Fraud_Label"].sum()
plt.hist(tr_fr)
```

```python
card_mean=df.groupby("Card_Type")["Card_Age"].mean()
plt.plot(card_mean)
```

```python
trans_type_avg=df.groupby("Transaction_Type")["Account_Balance"].mean()
plt.bar(trans_type_avg.index,trans_type_avg.values)
```

```python
prod_risk_avg=df.groupby("Merchant_Category")["Risk_Score"].mean()
plt.pie(prod_risk_avg.values,labels=prod_risk_avg,autopct="%1.1f%%")
```
