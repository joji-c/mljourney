```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("C:\\Users\\jojic\\OneDrive\\Desktop\\mljourney\\matplotlib-works\\penguin-casestudy\\penguin_data.csv")
```

```python
print(df.columns)
print(df.shape)
print(df.isnull().sum())
```

```python
df["bill_length_mm"].fillna(df["bill_length_mm"].mean(),inplace=True)
print(df.isnull().sum())
```

```python
df["bill_depth_mm"].fillna(df["bill_depth_mm"].mean(),inplace=True)
print(df.isnull().sum())
```

```python
df["flipper_length_mm"].fillna(df.groupby("sex")["flipper_length_mm"].transform("mean"),inplace=True)
print(df.isnull().sum())
```

```python
df["sex"].fillna(df["sex"].mode()[0],inplace=True)
print(df.isnull().sum())
```

```python
df["body_mass_g"].fillna(df["body_mass_g"].mean(),inplace=True)
df.isnull().sum()
```

```python
# correlation between bill length and body mass
plt.plot(df.groupby("bill_length_mm")["body_mass_g"].mean())
```

```python
# correlation between bill depth and body mass
plt.plot(df.groupby("bill_depth_mm")["body_mass_g"].mean())
```

```python
# correlation between flipper length and body mass
plt.plot(df.groupby("flipper_length_mm")["body_mass_g"].mean())
```

```python
# island wise body mass index
island_mass=df.groupby("island")["body_mass_g"].mean()
plt.bar(island_mass.index,island_mass.values)
```

```python
# species wise body mass
species_mass=df.groupby("species")["body_mass_g"].mean()
plt.bar(species_mass.index,species_mass.values)
```

```python
# body mass outliers
mass=df["body_mass_g"]
plt.boxplot(mass)
```
