```python
df.drop(columns=["Precautions","Treatment","Emergency"],inplace=True)
```

```python
df.columns
```

```python
symtoms_colums=["Symptom 1","Symptom 2","Symptom 3","Symptom 4","Symptom 5"]
df["all_symtoms"]=df[symtoms_colums].values.tolist()
df.drop(columns=symtoms_colums,inplace=True)
df.head() 
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
mlb=MultiLabelBinarizer()
symtom_encoded=mlb.fit_transform(df["all_symtoms"])
drop_cols=symtoms_colums.copy()
drop_cols.append("all_symtoms") 
```

```python
drop_cols
```

```python
symtom_encoded.dtype
```

```python
df.drop(columns="all_symtoms",inplace=True)
```

```python
symtom_encoded_df=pd.DataFrame(symtom_encoded,columns=mlb.classes_,index=df.index)
```

```python
df=pd.concat([df,symtom_encoded_df],axis=1)
df.head()
```

```python
df["disease"]=df["Disease"]
df.drop(columns="Disease",inplace=True)
df.head()
```

```python
x=df.iloc[:,1:-1]
y=df.iloc[:,-1]
```

```python
df["Dog Breed"].unique()
```

```python
"""from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct=ColumnTransformer([("encoder",OneHotEncoder(),["Dog Breed"])],remainder="passthrough")
x=pd.DataFrame(ct.fit_transform(x))  """
x.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
y=lb.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
df["disease"].value_counts()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,f1_score
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
print("accuracy_score",accuracy_score(y_test,y_pred)*100)
#print("f1_score",f1_score(y_test,y_pred)*100)
```

```python
from sklearn.naive_bayes import GaussianNB
gb=GaussianNB()
gb.fit(x_train,y_train)
y_pred=gb.predict(x_test)
print("accuracy_score",accuracy_score(y_test,y_pred)*100)
#print("f1_score",f1_score(y_test,y_pred)*100) 
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion="entropy",max_depth=4)
dt.fit(x_train,y_train)
y_pred = dt.predict(x_test)
print("accuracy_score",accuracy_score(y_test,y_pred)*100)
#print("f1_score",f1_score(y_test,y_pred)*100) 
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(random_state=1)
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
print("accuracy_score",accuracy_score(y_test,y_pred)*100)
#print("f1_score",f1_score(y_test,y_pred)*100) 
```
