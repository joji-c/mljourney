```python
import seaborn as sns
import pandas as pd
import numpy as np
df=pd.read_excel("C:\\Users\\jojic\\OneDrive\\Desktop\\mljourney\\seaborn-works\\flight-casestudy\\Data_Train.xlsx")
print(df.head(3))
print(df.columns)
df.info()
```

```python
df.isnull().sum()
```

```python
#cleaning date 
df["day"]=df["Date_of_Journey"].str.split("/").str[0]
df["month"]=df["Date_of_Journey"].str.split("/").str[1]
df["year"]=df["Date_of_Journey"].str.split("/").str[2]
```

```python
df.head()
```

```python
#deleting a column
df.drop(columns="Date_of_Journey",inplace=True)
```

```python
df["day"]=df["day"].astype("int")
df["month"]=df["month"].astype("int")
df["year"]=df["year"].astype("int")
df.info()
```

```python
df["departure_hr"]=df["Dep_Time"].str.split(":").str[0]
df["departure_min"]=df["Dep_Time"].str.split(":").str[1]
df.head()
```

```python
df.drop(columns="Dep_Time",inplace=True)
```

```python
df["departure_hr"]=df["departure_hr"].astype("int")
df["departure_min"]=df["departure_min"].astype("int")
df.head()
```

```python
df["Arrival_Time"]=df["Arrival_Time"].str.split(" ").str[0]
df["Arrival_hr"]=df["Arrival_Time"].str.split(":").str[0]
df["Arrival_min"]=df["Arrival_Time"].str.split(":").str[1]
df.drop(columns="Arrival_Time",inplace=True)
df.head()
```

```python
df["Total_Stops"].unique()
```

```python
df["Total_Stops"].value_counts()
```

```python
#changing null value 
df["Total_Stops"]=df["Total_Stops"].map({"non-stop":0,"1 stop":1,"2 stops":2,"3 stops":3,"4 stops":4,np.nan:1})
df.isnull().sum()
```

```python
df["Total_Stops"].value_counts()
```

```python
df["Duration_hr"]=df["Duration"].str.split(" ").str[0]
df["Duration_hr"]=df["Duration_hr"].str.split("h").str[0]

df["Duration_min"]=df["Duration"].str.split(" ").str[1]
df["Duration_min"]=df["Duration_min"].str.split("m").str[0]

df.drop(columns="Duration",inplace=True)
df.head()
```
