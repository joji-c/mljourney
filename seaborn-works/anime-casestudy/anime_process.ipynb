```python
import pandas as pd
import numpy as np
import seaborn as sns
df=pd.read_csv("C:\\Users\\jojic\\OneDrive\\Desktop\\mljourney\\seaborn-works\\anime-casestudy\\anime.csv")
print(df.head())
print(df.columns)
df.isnull().sum()
```

```python
df.head()
```

```python
df["start_date"].fillna(df["start_date"].mode()[0],inplace=True)
df["end_date"].fillna(df["end_date"].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
df["start_year"]=df["start_date"].str.split("-").str[0]
df["start_month"]=df["start_date"].str.split("-").str[1]
df["start_day"]=df["start_date"].str.split("-").str[2]
df.drop(columns="start_date",inplace=True)
df["start_year"]=df["start_year"].astype("int")
df["start_month"]=df["start_month"].astype("int")
df["start_day"]=df["start_day"].astype("int")
df.head()
```

```python
df["end_year"]=df["end_date"].str.split("-").str[0]
df["end_month"]=df["end_date"].str.split("-").str[1]
df["end_day"]=df["end_date"].str.split("-").str[2]
df.drop(columns="end_date",inplace=True)
df["end_year"]=df["end_year"].astype("int")
df["end_month"]=df["end_month"].astype("int")
df["end_day"]=df["end_day"].astype("int")
df.head()
```
