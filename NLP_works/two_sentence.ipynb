```python
import nltk
import numpy
import re
from nltk.corpus import stopwords
sw=stopwords.words("english")
```

```python
s1="Football worldcup is conducted by fifa every four years"
s2="Germany has won worldcup four times"
```

```python
s1=s1.lower()
s2=s2.lower() 
```

```python
s1=re.sub("[^a-zA-Z0-9 ]","",s1)
s2=re.sub("[^a-zA-Z0-9 ]","",s2)
```

```python
s1=nltk.word_tokenize(s1)
s2=nltk.word_tokenize(s2)
```

```python
filtered_s1=[n for n in s1 if n not in sw]
filtered_s2=[n for n in s2 if n not in sw]
print(filtered_s1)
print(filtered_s2) 
```

```python
#nltk.download("wordnet")
from nltk.stem import WordNetLemmatizer
lem=WordNetLemmatizer()
lem_s1=[]
lem_s2=[]
for c in filtered_s1:
    lem_s1.append(lem.lemmatize(c))
for c in filtered_s2:
    lem_s2.append(lem.lemmatize(c))
print(lem_s1)
print(lem_s2)
```

```python
sentence=[" ".join(lem_s1)," ".join(lem_s2)]
sentence
```
