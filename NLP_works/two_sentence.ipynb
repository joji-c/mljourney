```python
import nltk
import numpy
import re
```

```python
from nltk.corpus import stopwords
sw=stopwords.words("english")
```

```python
s1="Football worldcup is conducted by fifa every four years"
s2="Germany has won worldcup four times"
```

```python
s1=s1.lower()
s2=s2.lower() 
```

```python
s1=re.sub("[^a-zA-Z0-9 ]","",s1)
s2=re.sub("[^a-zA-Z0-9 ]","",s2)
```

```python
new_s1=nltk.word_tokenize(s1)
new_s2=nltk.word_tokenize(s2)
```

```python
cleaned_s1=[n for n in new_s1 if n not in sw]
print(cleaned_s1)
cleaned_s2=[n for n in new_s2 if n not in sw]
print(cleaned_s2) 
```

```python
from nltk.util import ngrams
gram_s1=ngrams(sequence=cleaned_s1,n=1)
token_s1=list(gram_s1)
gram_s2=ngrams(sequence=cleaned_s2,n=1)
token_s2=list(gram_s2)
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
ss1=[]
for t in token_s1:
    print(f"{t[0]}:{ps.stem(t[0])}")
    ss1.append(ps.stem(t[0]))
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
ss2=[]
for t in token_s2:
    print(f"{t[0]}:{ps.stem(t[0])}")
    ss2.append(ps.stem(t[0]))
```

```python
print(ss1)
print(ss2)
```
