```python
import nltk 
import re

nltk.download("stopwords")

from nltk.corpus import stopwords
sw=stopwords.words("english")
print(sw)
```

```python
sentence="India won their third t20 world cup. sanju samson was rewarded with player of the tournament"
sentence=sentence.lower()
new_sentence=[w for w in sentence.split(" ") if w not in sw]
new_sentence=" ".join(new_sentence)
cleaned_data=re.sub("[^a-zA-Z0-9 ]","",new_sentence)
print(cleaned_data)
```

```python
nltk.download("punkt_tab")
text="the sun rises in the east"

tokens=nltk.word_tokenize(text)
print(tokens)

filtered_tokens=[t for t in tokens if t not in sw]
print(filtered_tokens)
```

```python
sentence="the# jupyter$ is@ the second@ largest@@ planet"
sentence=sentence.lower()
sentence=re.sub("[^a-zA-Z0-9 ]","",sentence)
print(sentence)

tokens=nltk.word_tokenize(sentence)
filtered_tokens=[t for t in tokens if t not in sw]
filtered_sentence=" ".join(filtered_tokens)
print(filtered_sentence)
```

```python
from nltk.util import ngrams
grams=ngrams(sequence=filtered_tokens,n=1)
tokens=list(grams)
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
for t in tokens:
    print(f"{t[0]}:{ps.stem(t[0])}")
```

```python
from nltk.stem import SnowballStemmer
ss=SnowballStemmer(language="english")
for t in tokens:
    print(f"{t[0]}:{ss.stem(t[0])}")
```
