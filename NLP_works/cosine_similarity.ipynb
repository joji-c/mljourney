```python
import numpy as np
```

```python
s1 = [0.0375,0,-0.025,0.0375,0,0.0375,0.0375,0,0]
s2 = [0,0,-0.02,0,0,0,0,0.06,0.06]
```

```python
s1 = np.array([0.03,0,-0.25,0.03,0,0.03,0.03,0,0])
s2 = np.array([0,0,-0.02,0,0,0,0,0.6,0.6])
```

```python
#dot product
dot_product=np.dot(s1,s2)
print(dot_product)
```

```python
#magnitude
m_s1=np.linalg.norm(s1)
m_s2=np.linalg.norm(s2) 
print(m_s1)
print(m_s2)
```

```python
cos_similarity = dot_product/(m_s1 * m_s2)
print(cos_similarity)
```
