```python
s1 = "Machine learning helps computers learn patterns from data"
s2 = "Data science uses machine learning to analyze the data"
```

```python
s1 = s1.lower()
s2 = s2.lower()
```

```python
from nltk.corpus import stopwords
sw = stopwords.words("english")
```

```python
cleaned_sentence1 = [w for w in s1.split(" ") if w not in sw]
cleaned_sentence2 = [w for w in s2.split(" ") if w not in sw]
cleaned_sentence1 = " ".join(cleaned_sentence1)
cleaned_sentence2 = " ".join(cleaned_sentence2)
print(cleaned_sentence1)
print(cleaned_sentence2)
```

```python
import nltk
tokens_s1 = nltk.word_tokenize(cleaned_sentence1)
tokens_s2 = nltk.word_tokenize(cleaned_sentence2)
print(tokens_s1)
print(tokens_s2)
```

```python
from nltk.stem import WordNetLemmatizer
lematizer = WordNetLemmatizer()
lem1 = [lematizer.lemmatize(t) for t in tokens_s1]
lem2 = [lematizer.lemmatize(t) for t in tokens_s2]
print(lem1)
print(lem2)
```

```python
sentence = " ".join(lem1)," ".join(lem2)
print(sentence)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
num_values = vectorizer.fit_transform(sentence)
print(num_values.toarray())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(num_values[0],num_values[1])
similarity
```
