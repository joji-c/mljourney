```python
df.columns
df_copy=df.copy()
```

```python
#fixin nulls
df_copy.fillna({"Age":df_copy["Age"].median()},inplace=True)
df_copy.fillna({"Salary":df_copy["Salary"].mean()},inplace=True)
df_copy.fillna({"Gender":"unknown"},inplace=True)
df_copy.fillna({"SatisfactionScore":df_copy["SatisfactionScore"].mode()[0]},inplace=True)
```

```python
df_copy.isnull().sum()
```

```python
df_copy[df_copy["CustomerID"].duplicated()].shape
```

```python
df_copy=df_copy.drop_duplicates(subset=["CustomerID"],keep="first")
```
