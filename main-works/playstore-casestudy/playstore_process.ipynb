```python
#create a copy of the original data
df_copy=df.copy()
```

```python
#check all the non int values
df_copy[~df_copy["Reviews"].str.isnumeric()]
```

```python
#drop the str value as its only one and doesnt effect the overall calcullation
df_copy.drop(df_copy.index[10472],inplace=True)
```

```python
df_copy[~df_copy["Reviews"].str.isnumeric()]
```

```python
df_copy["Reviews"]=df_copy["Reviews"].astype("int")
```

```python
#now we clean the size
df_copy[~df_copy["Size"].str.isnumeric()]
```

```python
#replace M with 000000 and K with 000 in reviews
df_copy["Size"]=df_copy["Size"].str.replace("M","000")
df_copy["Size"]=df_copy["Size"].str.replace("k","")
df_copy["Size"]=df_copy["Size"].replace("Varies with device",np.nan)
df_copy.head()
```

```python
df_copy["Size"]=df_copy["Size"].astype("float")
```

```python
#now change price sign
df_copy[~df_copy["Price"].str.isnumeric()]
```

```python
df_copy["Price"]=df_copy["Price"].str.replace("$","")
```

```python
df_copy["Price"]=df_copy["Price"].astype("float")
```

```python
#check installs
df_copy["Installs"].unique()
```

```python
#remove , and + from installs
df_copy["Installs"]=df_copy["Installs"].str.replace("+","")
df_copy["Installs"]=df_copy["Installs"].str.replace(",","")
```

```python
df_copy["Installs"]=df_copy["Installs"].astype("int")
```

```python
df_copy["Installs"].unique()
```

```python
#check last updated
df_copy["Last Updated"].unique()
```

```python
#changing the date type to date time
df_copy["Last Updated"]=pd.to_datetime(df_copy["Last Updated"])
df_copy["Last Updated"].unique()
```

```python
df_copy["last_updated_year"]=df_copy["Last Updated"].dt.year
df_copy["last_updated_month"]=df_copy["Last Updated"].dt.month
df_copy["last_updated_day"]=df_copy["Last Updated"].dt.day
df_copy.drop(columns="Last Updated",inplace=True)
df_copy.head()
```

```python
#checking for duplicates
df_copy.shape
```

```python
df_copy[df_copy["App"].duplicated()].shape
```

```python
df_copy=df_copy.drop_duplicates(subset=["App"],keep="first")
```
